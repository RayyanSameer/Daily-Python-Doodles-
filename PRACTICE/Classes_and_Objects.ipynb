```python
"""Create a Class for a Car
Define a Car class with attributes like brand, model, and price. Create multiple car objects and display their details."""

class Car():
    def __init__(self,brand,model,price):
        self.brand = brand
        self.model = model
        self.price = price

    def  display_details(self):
        print(f"Brand : {self.brand}")
        print(f"Model: {self.model}")
        print(f"Price : {self.model}")

#Objects 

car1 = Car("Toyota","Fortuner", 5200000)
car2 = Car("Honda","Elevate", 1000000)   
car3 = Car("Mahindra","Scorpio",1500000) 

#Display details
print("Car 1 :")
car1.display_details()

print("Car 2 :")
car2.display_details()

print("Car 3 :")
car3.display_details()
```
