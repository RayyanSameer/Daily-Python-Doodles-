```python
#Create a car class that stores brand, model and year with a method age that calculates its age 

class Car:
    def __init__(self, model,brand,year):
        self.model = model
        self.brand = brand
        self.year = year

    @property
    def age(self):
        return 2024 - self.year

    def age_check(self):
        age = 2024 - self.year
        print(f"The car is {age} years old")

    def scrappable(self):
        if self.age >= 15:
            print("Vehicle is eligible for scrapping")
        else:
            print("Drivable")    


car1 = Car("Toyota","Fortuner",2010)
car2 = Car("Honda","Elevate", 2023)   
car3 = Car("Mahindra","Scorpio",2009) 


print("Car 1 : ")
car1.age_check()
car1.scrappable()
print("car 2 : ")
car2.age_check()
car2.scrappable()
print("Car 3 : ")
car3.age_check()
car3.scrappable()


        
```

```python
#Buid a students class marks and computes total and avg 

class Student:
    def __init__(self,student_id,student_name,english_grade,math_grade,science_grade):

        self.student_id = student_id
        self.student_name = student_name
        self.english_grade = english_grade
        self.math_grade = math_grade
        self.science_grade = science_grade
        
      

    def find_total(self):
                total = self.math_grade + self.science_grade + self.english_grade   

    def find_avg(self):
        total = self.find_total()
        average = total/3
        return average 

    def display_details(self):
        print(f"--- Student: {self.student_name} ({self.student_id}) ---")
        
        print(f"Total Marks: {self.find_total()}")
        
        print(f"Average Mark: {self.find_avg():.2f}") 
        print("------------")


s1 = Student("001","ABCD",65,46,21)
s2 = Student("002","DEFG",54,84,54)
s3 = Student("003","HIJK",64,45,32)

total_for_s1 = s1.find_total()
average_for_s1 = s1.find_avg()


print(f"Student 1's Total: {total_for_s1}")
print(f"Student 1's Average: {average_for_s1}")

print("\n--- Using the helper method for all students ---")

s1.display_details()
s2.display_details()
s3.display_details()

```
